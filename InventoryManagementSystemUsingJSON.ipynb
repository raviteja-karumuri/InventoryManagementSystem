```python
import json

def getDiscount(amount):
    if amount > 10000:
        return amount * 0.1
    elif amount > 5000:
        return amount * 0.05
    elif amount > 2000:
        return 50
    return 0

def savePurchaseTransaction(details):
    fd = open("Data/salejson.txt","w")
    fd.write(",".join(details)+'\n')
    fd.close()
    
def getGstPercent(productCategory):
    # Sample GST rates taken for calculation, not actual values
    match productCategory:
        case "Food":
            return 5;
        case "Electronics":
            return 18
        case default:
            return 28
    
def calculateTotal(total,productCategory):
    total = total - getDiscount(total)
    total = total * (1 +  getGstPercent(productCategory)/100)
    return total
    
def generateBill(products,purchaseDetails,quantity,remaining):
    productId = purchaseDetails[0]
    totalAmount=calculateTotal(products[productId]['Price']*quantity,products[productId]['Category'])
    print("--------------------BILL---------------------------")
    print("Name          ",products[productId]['Name'])
    print("Price         ",products[productId]['Price'],"Rs")
    print("Quantity      ",quantity)
    print("Billing amount",totalAmount,"Rs")
    print("--------------------------------------------------")
    
    # Updating the available quantity in the dictionary
    products[productId]['Qn'] = remaining
    # Convert the dictionary into string
    js= json.dumps(record)

    # Saving the record locally into a json file
    fd = open("Data/record.json","w")
    fd.write(js)
    fd.close()
    print("Thanks for purchasing")
    print("--------------------------------------------------")
    
    # Add purchase details to text file
    purchaseDetails.insert(1,str(totalAmount))
    purchaseDetails.insert(1,str(quantity))
    purchaseDetails.insert(1,str(products[productId]['Price']))
    purchaseDetails.insert(1,products[productId]['Name'])
    purchaseDetails.append(time.ctime())
    savePurchaseTransaction(purchaseDetails)
```

```python
import json
  
# Open the local json file in read mode
fd = open("Data/record.json","r")
js=fd.read()
fd.close()

record = json.loads(js)

print("------------------CONTENTS------------------------")
for key in record.keys():
    print(key,':',record[key]['Name'],"|",record[key]['Price'],"|",record[key]['Qn'])

print("--------------------------------------------------")
print()

ui_name = input("Enter your name          :")
ui_mail = input("Enter your email         :")
ui_phone = input("Enter your phone number :")
ui_id = input("Enter product Id           :")
ui_qn = int(input("Enter quantity required:"))

lst=[ui_id,ui_name,ui_mail,ui_phone]

print()
if ui_id in record.keys():
    if record[ui_id]['Qn'] <= 0:
        print("Sorry, We're are not having any inventory. Please visit again")
    elif ui_qn > record[ui_id]['Qn']:
        print("Sorry, We're are not having enough inventory.")
        ui_response = input("Do you want to proceed with purchase of " + str(record[ui_id]['Qn']) + " " + record[ui_id]['Name'] + " (Y/N):")
        if ui_response == "Y" or ui_response == "y":
            generateBill(record,lst,record[ui_id]['Qn'],0)
        else:
            print("Sorry for the inconvenience. Please visit again")
    else:
        generateBill(record,lst,ui_qn,record[ui_id]['Qn']-ui_qn)
else:
    print("Sorry! Product not found in the inventory")
```
